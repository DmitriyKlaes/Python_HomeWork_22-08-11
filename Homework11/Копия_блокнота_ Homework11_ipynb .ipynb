```python
import sympy

from sympy.plotting import plot
from sympy import symbols

x = symbols('x')

plot((-12)*sympy.sin(sympy.cos(x))*x**4 - 18 * x**3 + 5 * x**2 + 10 * x - 30, (x, -200, 200))

```

```python
import matplotlib.pyplot as plt
import numpy as np

x_mass = []
y_mass = []
massive = []
x = -200
while x < 200:
    y = -12 * x**4 * np.sin(np.cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30
    y_mass.append(y)
    x_mass.append(x)
    x += 0.1

plt.grid()
plt.plot(x_mass, y_mass)
plt.show()
```
